```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr

df = pd.read_csv("Dropout Rate vs. Juvenile Crime Rate by Year.csv")

pearson_corr, pp_val = pearsonr(df["Dropout Rate"], df["Juvenile Apprehended Rate"])
spearman_corr, sp_val = spearmanr(df["Dropout Rate"], df["Juvenile Apprehended Rate"])

print("Dropout Rate vs. Juvenile Crime Rate:")
print("Pearson Correlation:", round(pearson_corr, 3), "p-value:", round(pp_val, 4))
print("Spearman Correlation:", round(spearman_corr, 3), "p-value:", round(sp_val, 4))
```

```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr

df = pd.read_csv("Child Marriage, Child Trafficking and Dropout Rates by Year.csv")

pearson_cm, pval_cm = pearsonr(df["Dropout Rate"], df["Child Marriage Rate"])
spearman_cm, pval_cm_s = spearmanr(df["Dropout Rate"], df["Child Marriage Rate"])

print("Dropout vs Child Marriage:")
print("Pearson Correlation:", round(pearson_cm, 3), "p-value:", round(pval_cm, 3))
print("Spearman Correlation:", round(spearman_cm, 3), "p-value:", round(pval_cm_s, 3))

pearson_ct, pval_ct = pearsonr(df["Dropout Rate"], df["Child Trafficking Rate"])
spearman_ct, pval_ct_s = spearmanr(df["Dropout Rate"], df["Child Trafficking Rate"])

print("\nDropout vs Child Trafficking:")
print("Pearson Correlation:", round(pearson_ct, 3), "p-value:", round(pval_ct, 3))
print("Spearman Correlation:", round(spearman_ct, 3), "p-value:", round(pval_ct_s, 3))
```
